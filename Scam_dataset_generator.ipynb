```python
import json


def find_json(s):
    start = s.find("{")
    end = s.rfind("}")
    if start != -1 and end != -1:
        json_string = s[start : end + 1]
    else:
        json_string = "{{}}"
    # print(json_string)
    return json.loads(json_string)
```

```python
def get_user_scam_message(scam, tricks, attacks):

    return f"""
You are a helpful assistant which is helping me in generating a scam related dataset which in turn be used for good cause to detect scams.
Generate a Scam SMS of type '{scam}' in which scammer has used following tricks '{tricks}' and include following attacks '{attacks}'.

After you generate it, list reasons why it is a scam.

Just return the result in the following JSON schema (there MUST not be any other explanations):

{{
  "message": string,
  "reasons": [string]
}}
"""
```

```python
def get_user_non_scam_message(scam, tricks, attacks):

    return f"""
Generate a Normal SMS which might seems like a scam of type'{scam}', but it is not.
It also might have following tricks '{tricks}' but it is still not a scam message.
It also might seem like it contains following attacks '{attacks}' but it is still not a scam message.

After you generate it, list reasons why it should be a scam and also list reasons why it should not be a scam.
Based on that give final conclusion wheather it is a scam or not.

If it turns out to be a scam, update the output to be a non-scam.

Return the result in the following JSON schema:

Just return the result in the following JSON schema, no other explanation :

{{
  "message": string,
  "reasons_for_non_scam": [string]
}}
"""
```

```python
import pandas as pd
from itertools import permutations

MAX_OUTPUTS = 3
# START_FROM = 357
START_FROM = 1

data = []


def start_generating(frm=START_FROM, to=MAX_OUTPUTS):
    output_id = 0
    for i in range(1, 4):
        # print(i)
        if output_id >= MAX_OUTPUTS:
            break

        subScamsList = list(permutations(scams, i))
        # subScams = get_n_random_items(scams, i)

        for subScams in subScamsList:
            if output_id >= MAX_OUTPUTS:
                break

            scam_string = ", ".join(subScams)

            for j in range(1, 4):
                # print(j)
                if output_id >= MAX_OUTPUTS:
                    break

                subTricksList = list(permutations(tricks, j))

                for subTricks in subTricksList:
                    if output_id >= MAX_OUTPUTS:
                        break

                    trick_string = ", ".join(subTricks)

                    for k in range(1, 5):
                        # print(k)
                        if output_id >= MAX_OUTPUTS:
                            break

                        subAttacksList = list(permutations(attacks, k))

                        for subAttacks in subAttacksList:
                            if output_id >= MAX_OUTPUTS:
                                break

                            if output_id < START_FROM:
                                output_id += 1
                                print(output_id)
                                break

                            print(output_id)
                            output_id += 1

                            attack_string = ", ".join(subAttacks)

                            message = get_user_scam_message(
                                scam_string, trick_string, attack_string
                            )
                            print(message)

                            result, scam_reasons = getResponse(message)

                            data.append(
                                [
                                    result,
                                    scam_string,
                                    trick_string,
                                    attack_string,
                                    1,
                                    scam_reasons,
                                ]
                            )

                            message2 = get_user_non_scam_message2(
                                result, scam_string, trick_string, attack_string
                            )
                            result2, non_scam_reasons = getResponse(message2)

                            data.append(
                                [
                                    result2,
                                    scam_string,
                                    trick_string,
                                    attack_string,
                                    0,
                                    non_scam_reasons,
                                ]
                            )
```

```python
def save_to_file():
    df = pd.DataFrame(
        data,
        columns=[
            "content",
            "scam type",
            "trick type",
            "attack type",
            "is scam",
            "reason",
        ],
    )
    df.to_excel("scams13.xlsx", index=False)
```
